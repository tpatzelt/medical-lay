## inspect some TLC samples

```python
from config import TLCPaths
from models import Sample
from data_loading import annotations
```

```python
samples = []
for file in TLCPaths.json_dir.iterdir():
    sample = Sample.parse_file(file)
    samples.append(sample)


def get_sample_by_annotation_id(a_id):
    for s in samples:
        if any(a.id == a_id for a in s.annotations):
            return s
    return None#%%

def get_samples_by_mention(mention):
    ss = []
    for s in samples:
        if any(a.get_mention() == mention for a in s.annotations):
            ss.append(s)
    return ss
```

```python
stein_samples = get_samples_by_mention("Stein")
len(stein_samples)
```
